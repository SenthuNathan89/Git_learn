```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('acdc_biorithm.csv')
df = df.dropna(how='any',axis=0) 
df_efgr=df[df['type']=="eFGR"].reset_index(drop=True)
df_low=df[df['type']=="low risk"].reset_index(drop=True)
```

```python
df_efgr_ac=df_efgr.iloc[:,0:10]
df_efgr_ac=df_efgr_ac.round(3)
```

```python
df_low_ac=df_low.iloc[:,0:10]
df_low_ac=df_low_ac.round(3)
```
