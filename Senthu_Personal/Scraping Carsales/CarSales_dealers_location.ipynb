```python
import requests
name=[]
address=[]
number=[]
link=[]
N_pages=3
U=[]
```

```python
for i in range(1,N_pages):
    if i==1:
        URL = 'https://www.carsales.com.au/dealer-services/dealers?q=(And.Service.Carsales._.Market.Car._.Status.1._.State.Northern+Territory.)'

    elif i==2:
        URL = 'https://www.carsales.com.au/dealer-services/dealers?s=15&q=(And.Service.Carsales._.Market.Car._.Status.1._.State.Tasmania.)'

    else:
        URL = 'https://www.carsales.com.au/dealer-services/dealers?s=15&q=(And.Service.Carsales._.Market.Car._.Status.1._.State.Northern+Territory.)'
        n=str((i-1)*15)
        URL=URL[:54]+n+URL[56:]
    time.sleep(10)   
    user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1985.143 Safari/537.36'

    headers = {'User-Agent': user_agent}
    content = requests.get(URL)
    soup = BeautifulSoup(content.text, 'html.parser')
    
    if soup.find_all('h4')==[]:
        while soup.find_all('h4')==[]:
            content = requests.get(URL)
            soup = BeautifulSoup(content.text, 'html.parser')
            
        

    print(i)
    rows = soup.find_all('h4')
#     print(rows)
    for row in rows:          # Print all occurrences
#         print(len(row.get_text()))
        name.append(row.get_text())
        
    rows = soup.find_all('p')
    for row in rows:          # Print all occurrences
    #   print(row.get_text())
        address.append(row.get_text())

    for rating in soup.select("span.phone"):
#         print(len(rating.get("data-number")))
        number.append(rating.get("data-number"))
        
    for rating in soup.select('a.btn.btn-sm'):
        #print(rating.get("href"))
        link.append(rating.get("href"))
```

```python
address_street=[]
address_suburb=[]
for i in range (len(address)):
    if i==0:
        address_street.append(address[i])
    elif i==1:
        address_suburb.append(address[i])
    elif i%3==0:
        address_street.append(address[i])
    elif i%3==1:
        address_suburb.append(address[i])
        
```

```python
link_ref=[]
for i in range(len(link)):
    if link[i]!='#':
        link_ref.append(link[i])
```
