```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
from collections import Counter
```

```python
df_input=pd.read_csv('carsales_1_nt.csv')

name=[]
price=[]
Dealer=[]
D_name=[]
```

```python
# intialise data of lists. 
name=[]
price=[]
type_=[]
link_dealer=[]
name_dealer=[]

data = {'Car Name':name, 'Car Price':price, 'Car Type':type_, 'Dealer_Name':name_dealer, 'Dealer_Link':link_dealer} 
df = pd.DataFrame(data)
```

```python
df_input.shape[0]
```

```python

for i in range (df_input.shape[0]):
    x_state=0
    y_state=0
    
    U1="https://www.carsales.com.au"
    U2=df_input.Link[i]
#     print(U2)
    D=df_input.Name[i]
#     print(D)
    URL=U1+U2
    user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1985.143 Safari/537.36'

    headers = {'User-Agent': user_agent}
    content = requests.get(URL)
    soup = BeautifulSoup(content.text, 'html.parser')

    if str(soup.find(id="tab-new-cars"))=='None':
        x_state=1
#         print('Cant find new')
    else:
        div_new = soup.find(id="tab-new-cars")
    if str(soup.find(id="tab-used-cars"))=='None':
#         print('Cant find used')
        y_state=1
    else:
        div_used = soup.find(id="tab-used-cars")
        
    if x_state==1 and y_state==1:
        print("Both are unavailable")
        print(U2)
    new_=[]
    new_p=[]
    used_=[]
    used_p=[]
    n_c=[]
    u_c=[]

    if x_state==0:
        new=div_new.find_all('h4')
        new_price=div_new.find_all("div", class_="primary-price")

        for i in range(len(new)):
            new_.append(new[i].get_text())
            new_p.append(new_price[i].get_text())
        n_c=['new']*len(new_)

    if y_state==0:
        used=div_used.find_all('h4')
        used_price=div_used.find_all("div", class_="primary-price")

        for i in range(len(used)):
            used_.append(used[i].get_text())
            used_p.append(used_price[i].get_text())
        u_c=['used']*len(used_)

    name=new_+used_
    price=new_p+used_p
    type_=n_c+u_c
    link_dealer=[U2]*len(type_)
    name_dealer=[D]*len(type_)
    
    if x_state==1 and y_state==1:
        print("Both are unavailable")
        print(U2)
        name=['NA']
        price=['NA']
        type_=['NA']
        link_dealer=[U2]*len(type_)
        name_dealer=[D]*len(type_)
    else:
        name=new_+used_
        price=new_p+used_p
        type_=n_c+u_c
        link_dealer=[U2]*len(type_)
        name_dealer=[D]*len(type_)

    data = {'Car Name':name, 'Car Price':price, 'Car Type':type_, 'Dealer_Name':name_dealer, 'Dealer_Link':link_dealer} 
    df1 = pd.DataFrame(data)
    
    frames=[df,df1]
    df=pd.concat(frames)
    
```
